Check outlier on data

```python
def check_outliers(df_col,col_name):
  print(f"checking outliers on {col_name}")

  # Find Q1, Q3, IQR
  Q1 = df_col.quantile(0.25)
  Q3 = df_col.quantile(0.75)
  IQR = Q3 - Q1

  # Find Bottom Fence and Upper Fence
  boxplot_min = Q1 - 1.5 * IQR
  boxplot_max = Q3 + 1.5 * IQR

  #Show the calculation
  print('Q1:',Q1)
  print('Q3:',Q3)
  print('IQR:',IQR)
  print('Min (Lower inner fence):',boxplot_min)
  print('Max (Upper inner fence):',boxplot_max)

  # Filter value that <Bottom fence and >Upper fence
  filter_min = df_col < boxplot_min
  filter_max = df_col > boxplot_max

  # drop outlier using loc. So it will show only value exclude (~) bottom fence and Upper fence
  df_outlier = customers.loc[(filter_min | filter_max)]

  # Check data information
  df_outlier.count()

  print("Result:")
  if df_outlier.shape[0] == 0:
    print("no outliers found")
  else:
    print(f"there are {df_outlier.shape[0]} outliers found")
  print('')
```

# Retention Rate

```python
import datetime as dt
```

```python
df_retention = df.copy()
```

```python
# A function that will parse the date Time based cohort:  1 day of month
def get_month(x): return dt.datetime(x.year, x.month, 1)

# Create transaction_date column based on month and store in TransactionMonth
df_retention['transaction_month'] = df_retention['order_date'].apply(get_month)

# Grouping by customer_id and select the InvoiceMonth value
grouping = df_retention.groupby('customer_id')['transaction_month']

# Assigning a minimum InvoiceMonth value to the dataset
df_retention['cohort_month'] = grouping.transform('min')

# printing top 5 rows
print(df_retention.head())
```

```python
def get_date_int(df, column):
    year = df[column].dt.year
    month = df[column].dt.month
    day = df[column].dt.day
    return year, month, day

# Getting the integers for date parts from the `InvoiceDay` column
transcation_year, transaction_month, _ = get_date_int(df_retention, 'transaction_month')

# Getting the integers for date parts from the `CohortDay` column
cohort_year, cohort_month, _ = get_date_int(df_retention, 'cohort_month')
```

```python
#  Get the  difference in years
years_diff = transcation_year - cohort_year

# Calculate difference in months
months_diff = transaction_month - cohort_month


""" Extract the difference in months from all previous values
 "+1" in addeded at the end so that first month is marked as 1 instead of 0 for easier interpretation.
 """

df_retention['cohort_index'] = years_diff * 12 + months_diff  + 1
print(df_retention.head(5))
```

```python
df_retention.to_csv('cleaned_data1.csv', index=False)
```

```python
# Counting daily active user from each chort
grouping = df_retention.groupby(['cohort_month', 'cohort_index'])



# Counting number of unique customer Id's falling in each group of CohortMonth and CohortIndex
cohort_data = grouping['customer_id'].apply(pd.Series.nunique)
cohort_data = cohort_data.reset_index()

 # Assigning column names to the dataframe created above
cohort_counts = cohort_data.pivot(index='cohort_month',
                                 columns ='cohort_index',
                                 values = 'customer_id')

# Printing top 5 rows of Dataframe
cohort_data.head()
```

```python
print(cohort_counts.round(1))
```

```python
"""
Select the first column and store it to cohort_sizes as size of cohort === no. active customer in first day
We will store the 1st column as Cohort size i.e total Number of Customers in that Cohort
"""

cohort_sizes = cohort_counts.iloc[:,0]

"""
We will divide the values in other columns with Cohort Size in order to calculate the retention rate i.e Number of Customers in Each Cohort Index
"""

retention = cohort_counts.divide(cohort_sizes, axis=0)
```

```python
"""
this matrix has 2 axes:
x - cohort date (beginning of month); y - cohort maturity (in months)
every cell is a unique count of customer for the x and y

if a customer's last order was in period 2,
they were active in period 0, 1 as well
same for other periods

"""

# Coverting the retention rate into percentage and Rounding off.
retention.round(3)*100
```

## Visualizing the retantion rate

```python
# Initialize the figure
plt.figure(figsize=(16, 10))

# Adding a title
plt.title('Retention Rate in percentage:- Monthly Cohorts', fontsize = 14)

# Creating the seaborn based heatmap
sns.heatmap(retention, annot=True, fmt= '.0%',cmap='YlGnBu', vmin = 0.0 , vmax = 0.6)
plt.ylabel('Cohort Month')
plt.xlabel('Cohort Index')
# plt.yticks(rotation='360')
plt.show()
```

**Interpretation**

* Customers that has first purchase on May has a low retention on the next second and third month.

# RFM Segmentation

To do RFM segmentation, we need to identify which variables are R, F, and M

- **R**ecency: `recency`
- **F**requency: `total_sales` per account
- **M**onetary: `total_order` per account

Why do you think we should use these variables?

## Segmentation

```python
# Create RFM Score

rfm['r_score']  = pd.qcut(rfm['recency'], q = 5, labels = ['5','4','3','2','1'])
rfm['f_score'] = pd.qcut(rfm['frequency'], q = 5, labels = ['1','2','3','4','5'])
rfm['m_score'] = pd.qcut(rfm['monetary'], q = 5, labels = ['1','2','3','4','5'])
rfm['rfm_score'] = rfm['r_score'].astype('str') + rfm['f_score'].astype('str') + rfm['m_score'].astype('str')
rfm
```

## Customers Segment

```python
def customer_segmentation(segment):
    if segment in ['Champions', 'Loyal Customers', 'Customers Needing Attention', 'Potential Loyalist']:
        return 'Royal Customers'
    elif segment in ['At Risk', 'Can’t Lose Them']:
        return 'Absent Customers'
    elif segment in ['Recent Customers', 'Promising']:
        return 'Current Customers'
    else:
        return 'Sleeping Customers'
```

```python
rfm2['cust_seg'] = rfm2['segment'].apply(lambda x: customer_segmentation(x))
rfm2
```

```python
eda2_rfm = rfm2.groupby(['cust_seg']).agg({
    'recency' : ['mean'],
    'r_score' : ['mean'],
    'frequency' : ['mean'],
    'f_score' : ['mean'],
    'monetary' : ['mean'],
    'm_score' : ['mean'],
    'segment' : ['count']
})
eda2_rfm
```
